## Date Analysis Observations

1. The dataset covers a period from 03-Jan-2015 to 30-Dec-2018.
2. Sales transactions span approximately four years.
3. The earliest shipment occurred on 07-Jan-2015.
4. The latest shipment occurred on 05-Jan-2019.
5. Ship dates extend beyond the order date range because orders placed at the end of 2018 were delivered in early 2019.
6. The dataset is suitable for time-series analysis, seasonal trend analysis, and year-over-year performance comparisons.

```python
## Orders per year

df['Order Date'].dt.year.value_counts()
```

```python
df['Order Date'].dt.month.value_counts()
```

```python
df.groupby(df['Order Date'].dt.year)['Sales'].sum()
```
