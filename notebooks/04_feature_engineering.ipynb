```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
df = pd.read_csv("../data/processed/train_cleaned.csv",
parse_dates=['Order Date', 'Ship Date']
)
```

```python
df.head()
```

### --Order Year

1. Which year generated highest sales?

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df[['Order Date', 'Order Year']].head()
```

### Order Month 

2. Which month generates highest sales?

```python
df['Order Month'] = df['Order Date'].dt.month_name()
```

```python
df[['Order Month', 'Order Date']].head()
```

### Order Quater

3. Which quater perform best??

```python
df['Order Quarter'] = (
    'Q'  +
    df['Order Date'].dt.quarter.astype(str) 
)
```

```python
df[['Order Date','Order Quarter']].head()
```

### Shipping Speed Category

4. How many orders are Fast, Standard, or Slow?

```python
df['Shipping Speed']= np.where(
    df['Shipping Days'] <= 2,
    'Fast',
    np.where(
        df['Shipping Days'] <= 5,
        'Standard',
        'Slow'
    )
)
```

```python
df[['Shipping Days','Shipping Speed']].head()
```

### Sales Category

5. How many low-value and high-value orders exist?

```python
df['Sales Category'] = pd.cut(
    df['Sales'],
    bins=[0,100,500, 5000, 25000],
    labels=[
        'Low',
        'Medium',
        'High',
        'Permium'
    ]
)
```

```python
df[['Sales','Sales Category']].head()
```


### Customer Order Count

6.Who are repeat customers?

```python
customer_orders = (
    df.groupby('Customer ID')['Order ID']
    .nunique()
)

df['Customer Order Count'] = (
    df['Customer ID']
    .map(customer_orders)

)
```

### Customer Type

```python
df['Customer Type'] = np.where(
    df['Customer Order Count'] > 5,
    'Repeat Customer',
    'New Customer'
)
```

```python
df.head()
```

```python
df.to_csv(
    '../data/processed/feature_engineered_sales.csv',
    index=False
)
```
